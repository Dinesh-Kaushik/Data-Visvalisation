# Demographic Analysis:

```python
plt.figure(figsize=(8,8))
sns.barplot(data=df, x='Pclass', y='Age', hue='Gender', ci=None)
plt.title('Analysis of passenger acc. to their Age and Class')
plt.xlabel('types of class')
plt.ylabel('Age in no.')
plt.show()
```

```python
plt.figure(figsize=(9,8))
sns.scatterplot(data=df, x='Pclass',y='Age', hue='Gender')
plt.title('Age distribution among different classes')
plt.xlabel('class')
plt.ylabel('Age')
plt.show()
```

# Survival Analysis:

```python
plt.figure(figsize=(8,8))
sns.countplot(data=df,x='Survived')
plt.title('Analysis of survivers')
plt.xlabel('Survived')
plt.ylabel('No. of passenger')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.countplot(data=df, x='Survived', hue='Gender')
plt.title('Analysis of survivers')
plt.xlabel('Survived')
plt.ylabel('No. of passenger')
plt.show()
```

# Correlation Analysis:

```python
correlation_matrix =df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.show()
```

# Embarked and Fare Analysis:

```python
plt.figure(figsize=(9,3))
sns.countplot(data=df, x='Embarked',hue='Pclass')
plt.title('Distribution of passengers based on the port of embarkation')
plt.xlabel('Embarke')
plt.ylabel('No. of passenger')
plt.show()
```

```python
plt.figure(figsize=(10,3))
sns.barplot(data=df, x='Gender',y='Fare',hue='Pclass', ci=None)
plt.title('Relationship fare of with sex and class  ')
plt.xlabel('sex')
plt.ylabel('Fare in no.')
plt.show()
```

# Advanced Visualizations:

```python
plt.figure(figsize=(10,3))
sns.violinplot(data=df, x="Embarked", y="Survived",hue='Gender')
plt.title('Relationship of survivers based on the port of embarkation ')
plt.xlabel('Embarked')
plt.ylabel('Survived')
plt.show()
```
